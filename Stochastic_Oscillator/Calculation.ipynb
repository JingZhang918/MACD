```python
import yfinance as yf
data = yf.Ticker('600887.SS').history(start='2021-08-01', end='2021-10-01')[['Open','High','Low','Close']]
```

```python
data['L14'] = data.Low.rolling(window=14).min()
data['H14'] = data.High.rolling(window=14).max()
data['%K'] = (data.Close - data.L14)/(data.H14 - data.L14) * 100
data['%D'] = data['%K'].rolling(window=3).mean()
data['check_line'] = data['%D'].rolling(window=3).mean()
```

```python
import pandas_ta as ta
data.ta.stoch(high='High', low='Low', close='Close', k=14, d=3, append=True)
```
