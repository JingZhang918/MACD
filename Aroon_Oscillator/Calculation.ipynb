```python
import yfinance as yf
df = yf.Ticker('600887.SS').history(start='2021-08-01',end='2021-10-01')[['High','Low']]
```

```python
df.index = df.index.astype(int)
df = df.reset_index()

df['last_high'] = df.High.rolling(window=25).apply(lambda x: x.idxmax())
df['aroon_up'] = (25 - (df.index - df.last_high))/25 * 100
df['last_low'] = df.Low.rolling(window=25).apply(lambda x: x.idxmin())
df['aroon_down'] = (25 - (df.index - df.last_low))/25 * 100
df['aroon_oscillator'] = df.aroon_up - df.aroon_down

import time
df['Date'] = [time.strftime("%Y-%m-%d", time.localtime(ts/1e9)) for ts in df.Date]
df = df.set_index('Date')
```

```python
import pandas_ta as ta
df.ta.aroon(length=25, append=True)
```

```python
df.to_excel('aroon_oscillator_calculation.xlsx')
```
