```python
import yfinance as yf
import pandas_ta as pd
from plotly.subplots import make_subplots
import plotly.graph_objs as go
```

```python
df = yf.Ticker('600887.SS').history(start='2021-06-01', end='2021-10-01')[['Open','High','Low','Close']]
df.ta.aroon(length=25, append=True)
df.columns = [x.lower() for x in df.columns]
df.columns
```

```python
fig = make_subplots(rows=2, cols=1)
fig.append_trace(
    go.Candlestick(
        x=df.index,
        open=df.open,
        high=df.high,
        low=df.low,
        close=df.close,
        increasing_line_color='#ef476f',
        decreasing_line_color='#06d6a0',
        showlegend=False
    ), row=1,col=1
)
fig.append_trace(
    go.Scatter(
        x=df.index,
        y=df.aroonosc_25,
        line=dict(color='#ffd166', width=2),
        name='Aroon Oscillator',
        legendgroup='2',
    ), row=2,col=1
)
fig.append_trace(
    go.Scatter(
        x=df.index,
        y=df.aroonu_25,
        line=dict(color='#ef476f',width=2),
        legendgroup='2',
        name='Aroon Up'
    ), row=2, col=1
)
fig.append_trace(
    go.Scatter(
        x=df.index,
        y=df.aroond_25,
        line=dict(color='#06d6a0',width=2),
        legendgroup='2',
        name='Aroon Down'
    ), row=2, col=1
)


layout = go.Layout(
    plot_bgcolor='#efefef',
    font_family='Monospace',
    font_color='#073b4c',
    font_size=20,
    xaxis=dict(
        rangeslider=dict(visible=False)
    ),
    legend={'orientation':'h'
           }
)

fig.add_hline(y=0, col=1, row=2, line_color='#336699', line_width=2, line_dash='dash')



fig.update_xaxes(
    rangebreaks=[
        dict(bounds=['sat', 'mon'])
    ]
)

fig.update_layout(layout)
fig.show()
```
