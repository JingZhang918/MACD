```python
import yfinance as yf
df = yf.Ticker('600887.SS').history(start='2021-08-01', end='2021-10-01')[['Open','High','Close','Low']]
```

```python
import pandas as pd
df['HighMove'] = df.High.diff()
df['LowMove'] = -df.Low.diff()
df['PDM'] = df.HighMove.multiply((df.HighMove > df.LowMove) & (df.HighMove > 0).astype(int))
df['NDM'] = df.LowMove.multiply((df.LowMove > df.HighMove) & (df.LowMove > 0).astype(int))
df['TR'] = pd.concat([df.High-df.Low, df.High-df.Close.shift(1), df.Low-df.Close.shift(1)],axis=1).max(axis=1)
df['PDI'] = df.PDM.ewm(span=14).mean().div(df.TR.rolling(window=14).mean())*100
df['NDI'] = df.NDM.ewm(span=14).mean().div(df.TR.rolling(window=14).mean())*100
df['adx'] = 100*abs((df.PDI-df.NDI).div(df.PDI+df.NDI))
```
